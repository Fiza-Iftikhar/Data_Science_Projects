```python
movie['Gender'].unique()
```

```python
gender = movie.groupby("Gender")
```

```python
gender.groups
```

```python
male = gender.get_group('Male')
```

```python
female = gender.get_group('Female')
```

```python
male["Gender"].unique()
```

```python
female["Gender"].unique()
```

```python
male1 = male.iloc[:,3:9].sum()
female1 = female.iloc[:,3:9].sum()
df = pd.DataFrame({"male":male1, "female": female1})

ax = df.plot.bar(figsize = (10,7), color = ["#007acc","#2d8659"], rot = 0)
ax.set_facecolor(color = "#f2f2f2")
ax.legend(frameon = False)
ax.set_title("Most watched episode by Male and Female", fontsize = 15)
ax.tick_params(bottom = False, left = False)
for value in ax.spines.values():
    value.set_visible(False)

    
    
    
plt.savefig("Most watched episode by Male and Female.png")
plt.show()
```

```python
male2 = male.iloc[:,9:15].mean()
female2 = female.iloc[:,9:15].mean() 

df = pd.DataFrame({"male":male2, "female": female2})


ax = df.plot.bar(figsize = (10,7), color = ["#ac3973","#ff4da6"], rot = 0)
ax.set_facecolor(color = "#f2f2f2")
ax.legend(frameon = False)
ax.set_title("Most liked episode by Male and Female", fontsize = 15)
ax.tick_params(bottom = False, left = False)
for value in ax.spines.values():
    value.set_visible(False)



    
plt.savefig("Most liked episode by Male and Female.png")
plt.show()
```

### CONCLUSION
   - From analysis, we conclude that the most watched episode by Male and female is Episode-5. But the most liked episode by Male and Female is Episode-3.
