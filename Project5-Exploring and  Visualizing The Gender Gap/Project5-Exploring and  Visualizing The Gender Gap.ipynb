```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
```

```python
women_degrees.head()
```

```python
women_degrees["Year"].unique()
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 
             'Physical Sciences', 'Computer Science', 'Engineering']

lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 
                 'Art and Performance', 'Social Sciences and History']

other_cats = ['Health Professions', 'Public Administration', 'Education', 
              'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize = (18, 12))


for i in range(0,16,3):
    # print(i+1)
    ind = int(i/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees["Year"], women_degrees[stem_cats[ind]], label = "Women", color = "#66004d", marker = ".")
    ax.plot(women_degrees["Year"], 100- women_degrees[stem_cats[ind]], label = "Men", color = "#2d5986", marker = "*")
    ax.set_title(stem_cats[ind])
    ax.set_ylim(0,100)
    ax.set_xlim(1970,2012)
    ax.set_yticks([0,100])
    ax.tick_params(bottom = False, labelbottom = False)
    for value in ax.spines.values():
        value.set_visible(False)
    if i == 15:
        ax.tick_params(labelbottom = True)
    ax.legend()  
        
for j in range(1,16,3):
    # print(i+1)
    index = int(j/3)
    ax = fig.add_subplot(6,3,j+1)
    ax.plot(women_degrees["Year"], women_degrees[lib_arts_cats[index]], 
            label = "Women", color = "#66004d", marker = ".")
    ax.plot(women_degrees["Year"], 100- women_degrees[lib_arts_cats[index]],
            label = "Men", color = "#2d5986", marker = "*")
    ax.set_title(lib_arts_cats[index])
    ax.set_ylim(0,100)
    ax.set_xlim(1970,2012)
    ax.set_yticks([0,100])
    ax.tick_params(bottom = False, labelbottom = False)
    for value in ax.spines.values():
        value.set_visible(False)
    if index == 4:
        ax.tick_params(labelbottom = True)
    ax.legend()
        
for k in range(2,18,3):
    # print(i+1)
    inde = int(k/3)
    ax = fig.add_subplot(6,3,k+1)
    ax.plot(women_degrees["Year"], women_degrees[other_cats[inde]], 
            label = "Women", color = "#66004d", marker = ".")
    ax.plot(women_degrees["Year"], 100- women_degrees[other_cats[inde]],
            label = "Men", color = "#2d5986", marker = "*")
    ax.set_title(other_cats[inde])
    ax.set_ylim(0,100)
    ax.set_xlim(1970,2012)
    ax.set_yticks([0,100])
    ax.tick_params(bottom = False, labelbottom = False)
    for value in ax.spines.values():
        value.set_visible(False)
    if inde == 5:
        ax.tick_params(labelbottom = True)
    ax.legend()
    
plt.show()
```
